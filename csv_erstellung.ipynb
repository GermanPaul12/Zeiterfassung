```python
import pandas as pd
import numpy as np
import datetime as dt


def daterange(date1, date2):
    for n in range(int ((date2 - date1).days)+1):
        yield date1 + dt.timedelta(n)

dates = [dtt.strftime('%Y-%m-%d') for dtt in daterange(dt.datetime.strptime(f"{dt.datetime.now().year-2}-01-01", "%Y-%m-%d"),
                                                                                             dt.datetime.strptime("2025-09-30", "%Y-%m-%d"))]

kommen = [dt.datetime.strptime('08:00', '%H:%M') for i in range(len(dates))]
pause = [dt.datetime.strptime('0:30', '%H:%M') for i in range(len(dates))]
gehen = [dt.datetime.strptime('16:30', '%H:%M') if i % 3 != 0 else dt.datetime.strptime('15:30', '%H:%M') for i in range(len(dates))]   

weekday = [dt.datetime.strptime(date, '%Y-%m-%d').weekday() for date in dates]
dict_wochentag = {0:'Montag', 1:'Dienstag', 2:'Mittwoch', 3:'Donnerstag', 4:'Freitag', 5:'Samstag', 6:'Sonntag'}
wochentag = [dict_wochentag[i] for i in weekday]

dict_data = {'Datum':dates, 'Kommen':kommen, 'Pause':pause, 'Gehen':gehen}


df = pd.DataFrame(dict_data)
df['Wochentag'] = wochentag
df
```

```python
df['Summe'] = pd.to_timedelta(df.Gehen.strftime('%H:%M')) - pd.to_timedelta(df.Kommen.dt.strftime('%H:%M')) - pd.to_timedelta(df.Pause.dt.strftime('%H:%M'))
```
